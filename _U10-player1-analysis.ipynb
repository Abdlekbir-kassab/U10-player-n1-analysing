```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

# step 1: Creating the dataset with game-by-game performance

```python
data = {
    "Game": [1, 2, 3, 4, 5, 6, 7, 8],
    "Minutes": [20, 20, 20, 20, 20, 20, 20, 20],
    "Goals": [1, 0, 2, 1, 2, 1, 1, 2],
    "Assists": [1, 1, 0, 0, 1, 2, 2, 0],
    "Touches": [19, 18, 20, 19, 21, 25, 23, 26]
}

df = pd.DataFrame(data)
df
```

# step 2:  data cleaning 

```python
# Check for missing values
print("Any missing values?")
print(df.isnull().sum())

# Check data types
print("\nData types:")
print(df.dtypes)

# Quick preview of the data
print("\nPreview of the data:")
print(df.head())
```

# Step 4: Create Performance Metrics (for 20-min games)

```python
# Total goal contributions
df["Total_Contributions"] = df["Goals"] + df["Assists"]

# Touches per minute
df["Touches_per_Minute"] = df["Touches"] / df["Minutes"]

# Contributions per minute
df["Contributions_per_Minute"] = df["Total_Contributions"] / df["Minutes"]

# عرض الجدول المحدث
df
```

# Step 5: Descriptive Statistics

```python
df.describe()
```

# Step 6: Data Visualization

```python
# Plot 1: Goals, Assists, Touches per game

plt.figure(figsize=(12, 6))
plt.plot(df["Game"], df["Goals"], marker='o', label="Goals")
plt.plot(df["Game"], df["Assists"], marker='s', label="Assists")
plt.plot(df["Game"], df["Touches"], marker='^', label="Touches")

plt.title("Performance Over Games")

plt.xlabel("Game")
plt.ylabel("Count")

plt.legend()

plt.grid(True)
plt.tight_layout()

plt.show()

# Plot 2: Contributions per minute

plt.figure(figsize=(10, 5))
plt.bar(df["Game"], df["Contributions_per_Minute"], color="skyblue")

plt.title("Contributions per Minute (Per Game)")

plt.xlabel("Game")
plt.ylabel("Contributions per Minute")

plt.grid(axis='y')
plt.tight_layout()

plt.show()
```

# Step 7: Interpretation of Results

```python
- The player showed consistent involvement, with touches ranging from 18 to 26 per game.
- Goal contributions (goals + assists) were present in **all games**, showing strong offensive impact.
- The average contributions per minute is relatively high, considering each game lasts only 20 minutes.
- There’s a visible **increase in touches and assists** in later games, possibly showing improved confidence and connection with teammates.
- Game 5 and 7 showed peak all-around performance (both goals and assists).
```
